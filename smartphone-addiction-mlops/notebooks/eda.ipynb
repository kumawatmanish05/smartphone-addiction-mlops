# Data Cleaning and Preprocessing

```python
df = df.drop(columns=['transaction_id', 'user_id'])
```

```python
df['addiction_level'].value_counts()
```

```python
import pandas as pd

def fill_addiction_level(row):
    if pd.notnull(row['addiction_level']):
        return row['addiction_level']
    
    # Behavior-based logic
    if row['daily_screen_time_hours'] < 3:
        return 'Mild'
    elif row['daily_screen_time_hours'] < 6:
        return 'Moderate'
    else:
        return 'Severe'

df['addiction_level'] = df.apply(fill_addiction_level, axis=1)
```
